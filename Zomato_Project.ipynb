# **1. Restaurant Recommendation:** Build a recommendation system to suggest restaurants to users based on their preferences, such as cuisine type, location, and average cost.

**Filter the restaurants based on cuisine, location and average cost.**

```python
def get_recommendations(cuisine, area, avg_cost, top_n=5):
    filter_restaurants = data[(data['cuisines type'] == cuisine) & (data['area'] == area) &
        (data['avg cost (two people)'] <= avg_cost)]
    sort_restaurants = filter_restaurants.sort_values(by='avg cost (two people)')
    return sort_restaurants.head()
```

```python
user_cuisine = 'North Indian'
user_area = 'Indiranagar'
user_avg_cost = 300
recommendations = get_recommendations(user_cuisine, user_area, user_avg_cost)
print(recommendations[['restaurant name', 'cuisines type', 'area', 'avg cost (two people)']])
```

# **8. Customer Segmentation:** Use clustering techniques to segment customers based on their dining preferences and identify target customer groups for specific restaurants.

```python
data.columns
```

**Select the features based on Average cost and Number of Ratings**

```python
data.dropna(subset=['cuisines type', 'rate (out of 5)', 'avg cost (two people)', 'num of ratings'], inplace=True)
selected_features = ['rate (out of 5)', 'avg cost (two people)', 'num of ratings']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data[selected_features])
```

```python
n_clusters = 4
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
data['cluster'] = kmeans.fit_predict(data_scaled)
```

**Plot the Graph for Customer Segmentation based on Dining Preferences**

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=top_restaurants, x='rate (out of 5)', y='avg cost (two people)', palette='viridis')
plt.xlabel('Restaurant Rating')
plt.ylabel('Average Cost for Two People')
plt.title('Customer Segmentation based on Dining Preferences')
plt.legend(title='Cluster', title_fontsize='12')
plt.tight_layout()
plt.show()
```

The Graph shows Avereage Cost for Two People based on Restaurant Rating (out of 5).

**Define the minimum rating threshold and maximum cost threshold**

```python
min_rating_threshold = 4.5
max_cost_threshold = 1000
```

**Filter well-rated restaurants**

```python
well_rated_restaurants = data[data['rate (out of 5)'] >= min_rating_threshold]
```

**Filter restaurants with cost less than max_cost_threshold**

```python
affordable_well_rated_restaurants = well_rated_restaurants[well_rated_restaurants['avg cost (two people)'] <= max_cost_threshold]
```

**Select the top 5 restaurants based on average cost**

```python
top_5_restaurants = affordable_well_rated_restaurants.nsmallest(5, 'avg cost (two people)')
```

```python
average_costs = well_rated_restaurants.groupby(['restaurant type', 'area'])['avg cost (two people)'].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='restaurant name', y='avg cost (two people)', data=top_5_restaurants)
plt.title('Top 5 Well-Rated and Affordable Restaurants')
plt.xlabel('Restaurant Name')
plt.ylabel('Average Cost for Two People')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
plt.figure(figsize=(10, 6))
sns.scatterplot(x='restaurant name', y='area', data=top_5_restaurants, color= 'orange')
plt.xlabel('Restaurant Name')
plt.ylabel('Area of the Restaurants')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
